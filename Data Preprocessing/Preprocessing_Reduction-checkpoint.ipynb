Tuple Duplication: 

To identify number of duplicate tuples in the data:

```python
df_census.shape
```

```python
df_census.duplicated(subset=None, keep=False).value_counts()
```

Notes:

    duplicated() returns a boolean series
    subset=None considers all columns
    keep=False marks all duplicates as "True"
    Above dataset has 47 instances of duplicate tuples which can be removed

To remove duplicate tuples

```python
bool_series = df_census.duplicated(subset=None, keep=False)
```

```python
df_census_unique = df_census[~bool_series]
```

```python
print("Before removing duplicates:")
print(df_census.shape)
print("After removing duplicate tuples:")
print(df_census_unique.shape)
```

**DATA NORMALIZATION**

```python
data
```

```python
price_val = data['price']
price_val
```

```python
normalized_price = (price_val - price_val.min())/(price_val.max()-price_val.min())
data['normalized_price']=normalized_price
data.sort_values(by='price')
```

A Min-Max scaling is typically done via the following equation:

Xsc=Xâˆ’XminXmaxâˆ’Xmin.

```python
from sklearn import preprocessing
```

```python
x = data[['price']].values.astype(float)
x.max()
```

```python

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()
# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(data[['price']].values.astype(float))
# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)
data['normalized2']=x_scaled
data.sort_values(by='price')
```

**Z-Score Normalization**

```python
from scipy import stats
data['zscore']=stats.zscore(data['price'])
data.sort_values(by='price')
```
