Step3: Check for missing values
In Pandas missing data is represented by two value:

None: None is a Python singleton object that is often used for missing data in Python code.
NaN : NaN (an acronym for Not a Number), is a special floating-point value recognized by all systems that use the standard IEEE floating-point representation:

isnull()
notnull()
dropna()
fillna()
replace()
interpolate()

These could be scenarios of dropping the row:

Drop all rows that have any NaN (missing) values.

Drop only if the entire row has NaN (missing) values.

Drop only if a row has more than two NaN (missing) values.

Drop NaN (missing) in a specific column.

```python
# using imputer class
from sklearn.preprocessing import Imputer
data = np.array([[1, np.nan, 2], [2, 3, np.nan], [-1, 4, 2]])
```

```python
data
```

```python
imp = Imputer(strategy='mean')
imp.fit_transform(data)
```

```python
imp = Imputer(strategy='median')
imp.fit_transform(data)
```

```python
imp = Imputer(strategy='most_frequent')
imp.fit_transform(data)
```

Data Transformation: Categorical values

```python
import numpy as np
X = np.random.uniform(0.0, 1.0, size=(10, 2))
Y = np.random.choice(('Male','Female'), size=(10))
X[0]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
yt = le.fit_transform(Y)
```

```python
yt
```

```python
output = [1, 0, 1, 1, 0, 0]
```

```python
decoded_output = [le.classes_[i] for i in output]
```

```python
decoded_output
```

```python
#one-hot encoding using LabelBinarizer class
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
Yb = lb.fit_transform(Y)
```

```python
Yb
```

```python
lb.inverse_transform(Yb)
```
